```python
path = Path("experiments/neumann/20200504_blocks_8_10_100/validation_metrics/metrics.csv")
metrics_8 = pd.read_csv(path)
metrics_8.index = metrics_8.index.set_names(['epochs'])
metrics_8['Block'] = metrics_8["val_loss"].map(lambda x: "BLOCK_8")
```

```python
def base_chart(df, col_name, color_name):

    base = alt.Chart(
        df.reset_index(),
        width=300,
        height=200,
    ).mark_line(strokeWidth=5.0, point=True
    ).encode(
        x="epochs", 
        y=col_name,
        color=color_name,
        #shape="Block",
        #shape=alt.Shape('Block', scale=alt.Scale(range=['cross', 'circle', 'square', 'triangle-right', 'diamond']))
    )
    return base

def layer_charts(dfs, chart_func, col_name, color_name):
    """Return a layered chart."""

    alt_layer = alt.layer(*[chart_func(df, col_name, color_name) for df in dfs])
    return alt_layer

def concat_block_charts():
    ssim_chart = layer_charts([metrics_4, metrics_6, metrics_8], base_chart, "ssim", "Block")
    psnr_chart = layer_charts([metrics_4, metrics_6, metrics_8], base_chart, "psnr", "Block")
    concat_c = alt.vconcat(ssim_chart, psnr_chart)
#    concat_c = alt.vconcat(ssim_chart, psnr_chart, title="Experiments with different number of blocks")
    concat_c = concat_c.configure_axis(
        labelFontSize=12,
        titleFontSize=12,
    )
#     concat_c = concat_c.configure_legend(
#     titleFontSize=12,
#     labelFontSize=12,
#     strokeColor='gray',
#     fillColor='#EEEEEE',
#     padding=10,
#     cornerRadius=10,
#     orient='top-left'
#     )
    concat_c = concat_c.configure_title(fontSize=18) 
    return concat_c
    
```

```python
concat_block_charts()
```

# Unet

```python
path = Path("experiments/unet/20200504_unet_10_100/validation_metrics/metrics.csv")
unet_metrics = pd.read_csv(path)
unet_metrics = unet_metrics.drop(["psnr", "nmse"], axis=1)
unet_metrics.index = unet_metrics.index.set_names(['epochs'])
```

```python
unet_metrics.head()
```

```python
def metrics_chart(df, title):
    data = df.reset_index().melt('epochs')
    data.columns = ["epochs", "metric", "value"]
    chart = alt.Chart(data,
             width=600,
            height=300).mark_line(strokeWidth=3.0).encode(
        x='epochs',
        y='value',
        color='metric'
    )
    chart = chart.configure_axis(
            labelFontSize=20,
            titleFontSize=20,
        )
    chart = chart.configure_legend(
        strokeColor='gray',
        fillColor='#EEEEEE',
        padding=10,
        cornerRadius=10,
        orient='bottom-right'
    )
    chart = chart.properties(
        title=title,
    )
    chart = chart.configure_title(fontSize=24)  
#     chart = chart.configure_axis(
#         grid=False
#     )
    chart = chart.resolve_scale(
        color='independent'
    )    
    return chart
```

```python
metrics_chart(unet_metrics, "Unet mse, ssim, validation loss")
```

```python
path = Path("experiments/neumann/20200504_blocks_6_10_100/validation_metrics/metrics.csv")
metrics = pd.read_csv(path)
metrics = metrics.drop(["psnr", "nmse"], axis=1)
metrics.index = metrics.index.set_names(['epochs'])
metrics_chart(metrics, "Neumann network mse, ssim, validation loss")
```

```python
path = Path("experiments/unet/20200426_unet_epochs_50/validation_metrics/metrics.csv")
unet_metrics = pd.read_csv(path)
unet_metrics = unet_metrics.drop(["val_loss", "nmse", "ssim"], axis=1)
unet_metrics.index = unet_metrics.index.set_names(['epochs'])
```

```python
def chart_psnr(df, title):
    df = df.drop(["val_loss", "nmse", "ssim"], axis=1)
    df.index = unet_metrics.index.set_names(['epochs'])
    data = df.reset_index().melt('epochs')
    data.columns = ["epochs", "metric", "value"]
    chart = alt.Chart(data,
             width=600,
            height=300).mark_line(strokeWidth=3.0).encode(
        x='epochs',
        y='value',
        color='metric'
    )
    chart = chart.configure_axis(
            labelFontSize=20,
            titleFontSize=20,
        )

    chart = chart.properties(
        title=title,
    )
    chart = chart.configure_title(fontSize=24)  
    return chart
```

```python
path = Path("experiments/unet/20200504_unet_10_100/validation_metrics/metrics.csv")
metrics = pd.read_csv(path)
chart_psnr(metrics, "Unet PSNR")
```
